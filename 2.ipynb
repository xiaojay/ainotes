```python
def f(x):
    return 3*x**3+x**2 - 6*x +8
```

```python
def df(x):
    return 9*x**2+2*x-6
```

```python
x = 10
```

```python
lr = 0.1
```

```python
for i in range(100):
    y = f(x)
    gradient = df(x)
    print(x, y, gradient)
    
    x += -lr * gradient
```
