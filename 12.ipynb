```python
def f(x):
    return 3*x**2-9*x+7
```

```python
start = -2.0
stop = 2
step = 0.01
x = [start + i*step for i in range(int((stop-start)/step))]
```

```python
import random
y = [f(i) + 0.25*random.random() for i in x]
```

```python
import torch.nn as nn
```

```python
model = nn.Sequential(
    nn.Linear(1, 3),
    nn.ReLU(),
    nn.Linear(3, 1))
```

```python
loss = nn.MSELoss()
```

```python
import torch
```

```python
xt = torch.tensor([x,]).T
yt = torch.tensor([y,]).T
```

```python
trainer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(5000):
    pt = model(xt)
    l = loss(pt, yt)
    trainer.zero_grad()
    l.backward()
    trainer.step()
    print(f'epoch {epoch+1}, loss {l:f}')
```
