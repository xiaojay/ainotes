```python
def f(x):
    return 3*x+5
```

```python
def change(x):
    if abs(x) < 1:
        return x
    if x > 0:
        return 1
    else:
        return -1
```

```python
start = -100
stop = 100
step = 0.1
x = [start + i*step for i in range(int((stop-start)/step))]
```

```python
import random
y = [f(i) for i in x]
```

```python
len(x)
```

```python
data = list(zip(x, y))
```

```python
random.sample(data, 3)
```

```python
from micrograd.engine import Value
```

```python
len(data)
```

```python
def maxchange(x):
    while abs(x) > 10:
        x = x/10
    print(x)
    return x
```

```python
import random
a = 1; b= 2
a, b = [Value(a), Value(b)]
batch_size = 200
lr = 0.01

for i in range(100000):
    loss = None
    batch = random.sample(data, batch_size)

    for item in batch:
        x, y = item
        x = Value(x)
        y = Value(y)
        y_pred = a*x + b
        #print('pred:', y_pred.data, 'y', y.data)
        if not loss:
            loss = (y - y_pred)**2
        else:
            loss += (y - y_pred)**2
    loss = loss /batch_size
    #print('loss.grad', loss.grad)
    loss.backward()
    print('a.grad', a.grad, 'b.grad', b.grad,)
    a.data -= change(a.grad)
    b.data -= change(b.grad)
    a.grad = 0
    b.grad = 0

    print('a:', a.data, 'b:', b.data, 'loss:', loss.data)
```
