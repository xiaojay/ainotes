```python
from torch.utils.data import Dataset

class SimpleDataset(Dataset):
    def __init__(self, length=100):
        self.data = list(range(length))
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        # 数据点
        x = self.data[index]
        # 标签
        y = x ** 2
        return x, y
```
