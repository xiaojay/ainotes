```python
loss = nn.MSELoss()
```

```python
import torch
trainer = torch.optim.SGD(model.parameters(), lr=0.1)

for epoch in range(100):
    l = None
    for img, label in c2:
        x = img.view(-1).unsqueeze(0)
        pt = model(x)
        if label == 1:
            yt = torch.tensor([[1.0,0.0]])
        else:
            yt = torch.tensor([[0.0,1.0]])
        #print(pt)
        if not l:
            l = loss(pt, yt)
        else:
            l += loss(pt, yt)
        #print(l)
        
    l = l/len(c2)
    trainer.zero_grad()
    l.backward()
    trainer.step()
        
    print(f'epoch {epoch+1}, loss {l:f}')
```
