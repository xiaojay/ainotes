```python
loss = nn.MSELoss()
```

```python
trainer = torch.optim.SGD(model.parameters(), lr=0.05)
for epoch in range(100):
    for img, label in c2:
        x = img.view(-1).unsqueeze(0)
        pt = model(x)
        if label == 1:
            yt = torch.tensor([[1.0,0.0]])
        else:
            yt = torch.tensor([[0.0,1.0]])
        l = loss(pt, yt)
        trainer.zero_grad()
        l.backward()
        trainer.step()
    print(f'epoch {epoch+1}, loss {l:f}')
```
